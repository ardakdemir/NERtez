```python
data=open("2sizsufres.txt").readlines()
tags=["BPER","IPER","BLOC","ILOC","BORG","IORG"]
def tokenize(line):
    words=[]
    w=""
    for i in line:
        if i!="\t" and i!=" " and i!="\n":
            w+=i
        else:
            words.append(w)
            w=""
    if w!="":
        words.append(w)
    return words
#print(tokenize(data[0])[len(tokenize(data[0]))-1])
def fmeasure(pre,rec):
    return 2*pre*rec/(pre+rec)
def microfmeasure(tp1,fp1,tp2,fp2,fn1,fn2):
    pre=(tp1+tp2)/(tp1+tp2+fp1+fp2)
    rec=(tp1+tp2)/(tp1+tp2+fn1+fn2)
    return fmeasure(pre,rec)
def Conllaccuracy():
    count=0.0
    countsys=0.0
    true=0.0
    true2=0.0
    for line in data:
        tokens=tokenize(line)
        if tokens[len(tokens)-2] in tags :
            #print(line)
            if tokens[len(tokens)-1]!="O":
                true2+=1
                #countsys+=1
            if tokens[len(tokens)-1]==tokens[len(tokens)-2]:
                #print(line)
                true+=1
            else:
                o=1
                #print(line)
            count+=1
        if tokens[len(tokens)-1]!="O":
            countsys+=1
            if tokens[len(tokens)-2]=="O":
               # print(line)
                o=1
    print(true)
    print(count)
    pre=true/countsys
    rec=true/count
    ehbilen=true2/count
    fmes=fmeasure(pre,rec)
    return[fmes,ehbilen]
def Mucaccuracy():
    count=0.0
    countsys=0.0
    truetyp=0.0
    truetex=0.0
    for line in data:
        tokens=tokenize(line)
        if tokens[len(tokens)-2] in tags :
            #print(line)
            if tokens[len(tokens)-1][1:]==tokens[len(tokens)-2][1:]:##type
                #print(line)
                truetyp+=1
            if tokens[len(tokens)-1][0]==tokens[len(tokens)-2][0]:##text
                #print(line)
                truetex+=1
            count+=1
        if tokens[len(tokens)-1]!="O":
            countsys+=1
            
    pretyp=float(truetyp)/countsys
    rectyp=float(truetyp)/count
    pretex=float(truetex)/countsys
    rectex=float(truetex)/count
    ftyp=fmeasure(pretyp,rectyp)
    ftex=fmeasure(pretex,rectex)
    microf=microfmeasure(truetyp,countsys-truetyp,truetex,countsys-truetex,count-truetyp,count-truetex) 
    return [microf,ftyp,ftex]
def outfile():
    out=open("rootposresults.txt","w")
    out.write(str(count))
    out.write(str(true))
    out.write(str(true2))


[microf,ftyp,ftex]=Mucaccuracy()
[fmes,ehbilen]=Conllaccuracy()
print("Muc measure: %f"%ftex)
print("Conll measure: %f"%fmes)
```
